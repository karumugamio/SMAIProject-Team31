```
import wikipediaapi
import re
import requests
import wikipedia
import random

```

```
def getCategory(categoryMembers, level, max_level):
    categories=[]
    for c in categoryMembers.values():
        title = re.sub('Category:', '', c.title)
        categories.append(title)        
        if c.ns == wikipediaapi.Namespace.CATEGORY and level < max_level:
            getCategory(c.categorymembers, level=level + 1, max_level=max_level)
    return categories
```

```
def getCategoryPage(categories):
    catPages=[]
    for c in categories:
        S = requests.Session()
        URL = "https://en.wikipedia.org/w/api.php"
        PARAMS = {
            "action": "query",
            "cmtitle": "Category:"+c,
            "cmlimit": "20",
            "list": "categorymembers",
            "format": "json"
        }
        R = S.get(url=URL, params=PARAMS)
        DATA = R.json()
        PAGES = DATA['query']['categorymembers']
        for page in PAGES:
            catPages.append(page['title'])
            
    return catPages 
```

```
def getPageContent(title):
        try:
            page_content = wikipedia.page(title)
        except Exception as e:
            return "None"
        
        return(page_content.content)
```

```
def importPageContent(location, pageList):
    for pageTitle in pageList:
        content = getPageContent(pageTitle)
        if (content != None):
            file = location+"/"+pageTitle+".txt"
            try:           
                with open(file, 'w', encoding="utf8") as f:
                    f.write(content)
                    f.close()
            except FileNotFoundError:
                    print("no file")
```
